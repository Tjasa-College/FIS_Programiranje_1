Na fakulteti smo izmerili višino in starost študentov ter študentk. Študent,
ki je beležil rezultate ne zna programirati in jih je zapisal v tri sezname.
Potrebuje našo pomoč in sicer želi metodo Uredi, ki zbere podatke in jih
zloži v enotno podatkovno strukturo - slovar studenti. Potrebuje tudi
metode:</br>
Najvecji...poišče najvišjega med študenti</br>
Najmlajsi...poišče najmanjšega med študenti</br>
Povprecen...poišče tistega, ki je najbližje povprečni velikosti</br>
Izpis...izpiše podatke za študenta</br>
Podatki, ki so na voljo:</br>
ime = ["Janez","Miha","Ana","Angela","Rozi","Matevž","Pepca"]</br>
visina = [1.88, 1.91, 1.65, 1.77, 1.71, 1.75, 1.80]</br>
starost = [22, 23, 20, 21, 20, 21, 19]</br>


```python
ime = ["Janez", "Miha", "Ana", "Angela", "Rozi", "Matevž", "Pepca"]
visina = [1.88, 1.91, 1.65, 1.77, 1.71, 1.75, 1.80]
starost = [22, 23, 20, 21, 20, 21, 19]

def uredi(ime, visina, starost):
    studenti={}
    for i, v, s in zip(ime, visina, starost):
        studenti[i]={'visina':v, 'starost':s}
    return studenti

studenti=uredi(ime, visina, starost)
#print(studenti)

def najvecji(slovar):
    imena = list(slovar.keys())
    max_ime=imena[0]
    max_visina = slovar[max_ime]['visina']

    for kljuc in slovar:
        if slovar[kljuc]['visina'] > max_visina:
            max_visina = slovar[kljuc]['visina']
            max_ime = kljuc
    
    return max_ime, max_visina

def najmlajsi(slovar):
    imena = list(slovar.keys())
    min_ime=imena[0]
    min_starost = slovar[min_ime]['starost']

    for kljuc in slovar:
        if slovar[kljuc]['starost'] < min_starost:
            min_starost = slovar[kljuc]['starost']
            min_ime = kljuc
    return min_ime, min_starost


#print(povprecje)

povprecje = sum(visina) / len(visina)
def povprecen(slovar, povprecna):
    imena = list(slovar.keys())
    povp_ime = imena[0]
    povp_razlika = abs(slovar[povp_ime]['visina'] - povprecna)

    for kljuc in slovar:
        razlika = abs(slovar[kljuc]['visina'] - povprecna)
        if razlika < povp_razlika:
            povp_razlika = razlika
            povp_ime = kljuc

    return povp_ime, slovar[povp_ime]['visina']

def izpis(slovar, ime):
    if ime in slovar:
        v = slovar[ime]["visina"]
        s = slovar[ime]["starost"]
        print(f"{ime}:\nvišina = {v} m\nstarost = {s} let")
    else:
        print("Študenta ni v slovarju.")

print(najvecji(studenti))
print(najmlajsi(studenti))
print(povprecen(studenti, povprecje))
izpis(studenti, "Janez")
izpis(studenti, "Minka")
```
