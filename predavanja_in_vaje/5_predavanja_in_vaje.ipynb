```python
#Napišite lastno metodo Izracun v Pythonu, ki prepozna lik: kvadrat, trikotnik, krog, pravokotnik in vrne njegovo ploščino ter obseg. Napišite še metodo Izpis, ki vrnjeni
#vrednosti izpiše. Vse rezultate obvezno preverite
import math
def Izracun(lik, *mere):
    match lik.lower():
        case 'kvadrat':
            (a,)= mere
            ploscina = a * a
            obseg = 4 * a
        case 'trikotnik':
            (a, b, c,) = mere
            s = (a + b + c) / 2
            ploscina = (s*(s-a)*(s-b)*(s-c))**(1/2)
            obseg = a + b + c
        case 'krog':
            (r,) = mere
            ploscina = math.pi * r**2
            obseg = 2 * math.pi * r
        case 'pravokotnik':
            (a, b,) = mere
            ploscina = a * b
            obseg = 2*a + 2*b
        case _:
            return print('no no')
    return ploscina, obseg

def Izpis(lik, ploscina, obseg):
    print(f'Lik: {lik}')
    print(f'Ploščina: {ploscina}')
    print(f'Obseg: {obseg}')

p, o = Izracun('kvadrat', 4)
Izpis('kvadrat', p, o)

p, o = Izracun('pravokotnik', 3, 5)
Izpis('pravokotnik', p, o)

p, o = Izracun('krog', 2)
Izpis('krog', p, o)

p, o = Izracun('trikotnik', 3, 4, 5)
Izpis('trikotnik', p, o)
```

```python
#Napišite lastno funkcijo absolutno, ki z uporabo metode map() pretvori vrednosti elementov seznam v absolutne.
#Testni primer: seznam = [-1, -2, 0, 1, 2]
#Dodatek: poskusite rešiti nalogo brez uporabe funkcije in metode map() na čim krajši način

def absolutno(seznam):
    return list(map(abs, seznam))


seznam = [-1, -2, 0, 1, 2]
print(absolutno(seznam)) 
```
