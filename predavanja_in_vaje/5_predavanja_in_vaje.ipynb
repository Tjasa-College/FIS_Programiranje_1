```python
#primer uporabe **kwargs (po konvenciji)
#izračunajte vsoto danih treh spremenljivk:
#a = 1
#b = 2
#c = 3

def vsota(**kwargs):
    sum = 0
    for key, value in kwargs.items():
        sum += value
    return sum
print(vsota(a = 1, b = 2, c = 3))
```

```python
#Napišite lastno metodo Izpis v Pythonu, prebere parametre masa v kg, visina v centimetrih in barvaLas s pomočjo parametra **kwargs. Izpiše naj vrednosti vseh
#parametrov in kg pretvori v tone, visino v metre, barvo las pa izpiše z velikimi črkami.

#moja rešitev
def Izpis(**parametri):
    masa = parametri.get('masa')
    visina = parametri.get('visina')
    barva_las = parametri.get('barvaLas')

    if masa is not None:
        masa_t =masa / 1000 #iz kg v tone
        print(f"masa = {masa} kg =  {masa_t} T")
    if visina is not None:
        visina_m = visina/100 #iz cm v m
        print(f"visina = {visina} cm = {visina_m} m")
    if barva_las is not None:
        barva_las_velika = barva_las.upper()
        print(f"barvaLas = {barva_las} = {barva_las_velika}")   
    
    

    
Izpis(masa=66.5, visina=170, barvaLas="črni")

Izpis(masa=66.5)
```

```python
#njegova rešitev
def Izpis(**kwargs):
   for key, value in kwargs.items():
      match key:
         case 'masa':
            print(f'{key} = {value} kg = {value / 1000} T')
         case 'visina':
            print(f'{key} = {value} cm = {value / 100} m')
         case 'barvaLas':
            print(f'{key} = {value} = {value.upper()}')
kg = 66.5
cm = 170
lasje = 'črni'
Izpis( masa = kg, visina = cm, barvaLas = lasje)
```
