```python
"""
Nari코ite raztreseni grafikon za funkcijo y=sin(x) na intervalu [-2洧뤛, 2洧뤛] s korakom
洧뤛/8
Nasvet: poskrbite za pravilne oznake osi

"""
import matplotlib.pyplot as plt
import math

start = -2 * math.pi
end = 2 * math.pi
step = math.pi / 8

x = []
val = start
while val <= end + 1e-12:
    x.append(val)
    val += step

y = [math.sin(v) for v in x]

colors = list(range(len(x)))

plt.scatter(x, y, c=colors)
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("y = sin(x)")

plt.grid(False)
plt.show()

```

```python
"""
Na spletni strani Statisti캜nega urada RS poi코캜ite podatke o registriranih vozilih. Dopolnite tabelo in nari코ite ustrezen stolpi캜ni diagram in ga oblikujte. Vsak stolpec naj bo prikazan s svojo barvo
"""
import matplotlib.pyplot as plt

# Podatki (Vir: SURS, stanje 31. 12. 2024)
kategorije = ["Motorna kolesa", "Osebni avtomobili", "Avtobus", "Tovornjak", "Traktor"]
vrednosti = [95001, 1266663, 3015, 107014, 121956]

# Vsak stolpec druga barva (poljubno izbrane)
barve = ["tab:blue", "tab:orange", "tab:green", "tab:red", "tab:purple"]

plt.figure(figsize=(10, 5))
stolpci = plt.bar(kategorije, vrednosti, color=barve)

plt.title("Registrirana vozila v Sloveniji (stanje 31. 12. 2024)  Vir: SURS")
plt.xlabel("Vrsta vozila")
plt.ylabel("맚evilo registriranih vozil")

# Brez grida
plt.grid(False)

# Lep코i prikaz velikih 코tevil
plt.ticklabel_format(style='plain', axis='y')

# Prikaz vrednosti nad stolpci
for bar in stolpci:
    h = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        h,
        f"{int(h):,}".replace(",", "."),
        ha="center",
        va="bottom"
    )

plt.tight_layout()
plt.show()
```

```python
"""
Na kmetiji smo pre코teli 쬴vali. Rezultati so v spodnji tabeli. Nari코ite tortni grafikon, mu
dodajte naslov in legendo. Obarvajte posamezne vrednosti s svojimi barvami.
71
ma캜ke 3
pujsi 5
krave 4
konji 1
psi 2
"""
import matplotlib.pyplot as plt

oznake = ["ma캜ke", "pujsi", "krave", "konji", "psi"]
vrednosti = [3, 5, 4, 1, 2]

# Vsak kos svojo barvo
barve = ["tab:blue", "tab:orange", "tab:green", "tab:red", "tab:purple"]

plt.figure(figsize=(7, 6))

# Tortni grafikon
wedges, texts, autotexts = plt.pie(
    vrednosti,
    labels=None,            # oznake bomo dali v legendo
    colors=barve,
    autopct="%1.1f%%",      # procenti na grafu
    startangle=90
)

plt.title("콯ivali na kmetiji")

# Legenda (pove쬰 barve z oznakami)
plt.legend(
    wedges,
    oznake,
    title="Vrste 쬴vali",
    loc="center left",
    bbox_to_anchor=(1, 0.5)
)

plt.tight_layout()
plt.show()

```

```python
"""
1. Ustvarite datoteko z vi코inami 코tudentov in jo shranite.
2. Preberite datoteko z uporabo Numpy in izra캜unajte:
- povpre캜no vi코ino 코tudenta
- kdo je najvi코ji in kdo najni쬹i
- izpi코ite urejene podatke po velikosti
85
코tudent vi코ina
Tine 183
Tone 179
Jure 194
Meta 170
Anton 189
"""

import numpy as np

# preberemo: imena kot string, vi코ine kot int
imena, visine = np.loadtxt(
    "C:/Users/tjasa/Desktop/FIS_Programiranje_1/predavanja_in_vaje/Uvod_v_Python/visine.txt",
    dtype={"names": ("ime", "visina"), "formats": ("U20", "i4")},
    unpack=True
)

# povpre캜na vi코ina
povprecje = np.mean(visine)

# indeks najni쬵ega in najvi코jega
i_min = np.argmin(visine)
i_max = np.argmax(visine)

print(f"Povpre캜na vi코ina: {povprecje:.2f} cm")
print(f"Najni쬵i: {imena[i_min]} ({visine[i_min]} cm)")
print(f"Najvi코ji: {imena[i_max]} ({visine[i_max]} cm)")

# urejeni podatki po vi코ini (nara코캜ajo캜e)
urejeni_indeksi = np.argsort(visine)

print("\nUrejeno po vi코ini (nara코캜ajo캜e):")
for i in urejeni_indeksi:
    print(f"{imena[i]} {visine[i]}")

# (opcijsko) 캜e 쬰li코 padajo캜e:
print("\nUrejeno po vi코ini (padajo캜e):")
for i in urejeni_indeksi[::-1]:
    print(f"{imena[i]} {visine[i]}")
```
