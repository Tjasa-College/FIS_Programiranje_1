```python
#ploščina enakostraničnega trikotnika

a = 5
v = (a * 3**0.5)/2
p = (a * v)/ 2
print("Ploščina enakostraničnega trikotnika", p)
```

```python
# ploščina in obseg kroga
import math
d = 4

o = math.pi * d
r = d/2
p = math.pi * r**2

print("obseg in ploščina kroga",o, p)
```

```python
# ploščina in obseg enakokrakega trapeza
a = 8
b = 4
c = 6
d = 4

o = a + b + c + d
print("Obseg enakokrakega trapeza", o)

m = (a + c)/2
v = math.sqrt(b**2 - (1/4)*(a-c)**2)
p = m * v
print("Ploščina enakokrakega trapeza", p)
```

```python
#7 domet izstrelka
#kot izstrelka in hitrost izstrelka
import math
kot_izstrelka = int(input("Vnesi kot izstrelka v stopinjah: "))
hitrost_izstrelka = float(input("Vnesi hitrost izstrelka v m/s: "))
g = 9.81

radiani = math.radians(kot_izstrelka)
#l=v0^2/g *sin(alfa)
l = (hitrost_izstrelka**2 / g) * math.sin(radiani)
print("Domet izstrelka je:", round(l, 2), "metrov")
```

```python
#8.naloga: prebere trimestno število in izpiše vsoto števk

stevilo = int(input("Vnesite število: "))
originalno_stevilo = stevilo
sum = 0
while stevilo > 0:
    ostanek = stevilo % 10
    sum += ostanek
    stevilo = stevilo // 10

print(f"Vsota števk števila {originalno_stevilo} je {sum}")
```

```python
#9. naloga: iz col v mm 1 col (1″) = 25,4
cole = int(input("Vnesi število col: "))
print(f"{cole} col platišče je ", cole * 25.4, "mm")
```

```python
#naloga 10: diagonala

diagonala = int(input('vnesite diagonalo TV v colah: '))
razdalja = diagonala * 25.4
print (razdalja)
m = int(razdalja // 1000)
print('m = ', m)
dm = (int(razdalja-m*1000)//100)%100
print('dm = ', dm)
cm = (int(razdalja-m*1000-dm*100)//10)%10
print('cm = ', cm)
mm = int(razdalja-m*1000-dm*100-cm*10)
print('mm = ', mm)
```
