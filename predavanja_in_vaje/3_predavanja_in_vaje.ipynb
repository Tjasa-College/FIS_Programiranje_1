Atlet teče maraton, ki je dolg dva kroga. Napišite program, ki prebere čas
prvega in čas drugega kroga ter izpiše skupni čas atleta. Zazna naj vse
napačne vrednosti in izpiše ustrezno obvestilo. 

```python
print("=== 1. krog ===")
h1 = int(input("Ure: "))
m1 = int(input("Minute: "))
s1 = int(input("Sekunde: "))

print("=== 2. krog ===")
h2 = int(input("Ure: "))
m2 = int(input("Minute: "))
s2 = int(input("Sekunde: "))

# preverjanje pravilnosti
napaka = False

if h1 < 0 or h2 < 0:
    napaka = True
if m1 < 0 or m1 > 59 or m2 < 0 or m2 > 59:
    napaka = True
if s1 < 0 or s1 > 59 or s2 < 0 or s2 > 59:
    napaka = True

if napaka:
    print("NAPAKA: Vnešeni so bili neveljavni časi!")
else:
    # pretvorba v sekunde
    total_seconds = (h1*3600 + m1*60 + s1) + (h2*3600 + m2*60 + s2)

    # pretvorba nazaj
    h = total_seconds // 3600
    total_seconds %= 3600
    m = total_seconds // 60
    s = total_seconds % 60

    print(f"Skupni čas: {h} ur, {m} min, {s} s")
```

Napišite program, ki vnese telesno višino in maso, izračuna indeks telesne
mase ter ga izpiše in še ugotovitev glede indeksa. Nasvet: internet

```python
teza = float(input("Vnesi težo v kg: "))
visina = float(input("Vnesi višino v cm: "))

if teza >=0 or visina >=0:
    ITM = teza/(visina**2)
    if ITM < 18.5:
        print("Podhranjeni")
    elif 18.5 <= ITM <= 24.99: 
        print('Normalna teža')
    elif 25 <= ITM <= 29.99: 
        print('Povišana teža')
    elif 30 <= ITM <= 34.99: 
        print('Debelost 1')
    elif 35 <= ITM <= 40.99: 
        print('Debelost 2')
    else:
        print('Debelost 3')
else:
    print("Vpišite primerno težo in višino")
```

vsa števila med 100 in 1000 katerih vsota števk je manjša od 5 (npr. 121 = 1+2+1<5)

```python
for i in range(100, 1000):
    sum = 0;
    stevka = i

    while stevka != 0:
        ostanek = stevka % 10
        sum +=  ostanek
        stevka = stevka // 10
    
    if sum < 5:
        print(i)

```

vsa popolna števila med 1 in 1000 (število je enako vsoti vseh deliteljev, razen
samega sebe, npr. 6=3+2+1)


```python
for number in range(1, 1000):
    sum = 0
    for i in range(1, (number//2) +1):
        if number % i == 0:
            sum +=i
    if sum == number:
        print(number)
    
```
