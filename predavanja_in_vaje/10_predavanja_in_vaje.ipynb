```python
"""
V spletni učilnici (vaje) je datoteka zaposleni.csv.
Naloge:
a. preberite podatke iz datoteke zaposleni.csv v pandas
podatkovni okvir.
b. izpišite osnovne statistične podatke o plačah (povprečna,
najmanjša in največja plača).
c. preštejte število zaposlenih po posameznih stopnjah izobrazbe.
d. izračunajte povprečne plače po oddelkih.
e. poiščite zaposlenega z najvišjo plačo in izpišite njegove
podatke.

"""

import pandas as pd
df = pd.read_csv(
    r"C:/Users/tjasa/Desktop/FIS_Programiranje_1/predavanja_in_vaje/Uvod_v_Python/Zaposleni.csv",
    sep=";",          
    encoding="utf-8"  
)

df["Plača"] = (
    df["Plača"].astype(str)
      .str.replace("€", "", regex=False)
      .str.strip()
      .str.replace(".", "", regex=False)   
      .str.replace(",", ".", regex=False) 
      .astype(float)
)

print("Povprečna:", df["Plača"].mean())
print("Najmanjša:", df["Plača"].min())
print("Največja:", df["Plača"].max())

print("Število zaposlenih po izobrazbi:")
print(df["Izobrazba"].value_counts())

print("Povprečne plače po oddelkih:")
print(df.groupby("Oddelek")["Plača"].mean())

print("Zaposleni z najvišjo plačo:")
naj = df.loc[df["Plača"].idxmax()]
print(naj)
```

```python
"""
Doma imamo cisterno za vodo za zalivanje vrta. Cisterna je v obliki valja
in izmerimo ji lahko premer ter višino. S pomočjo razredov ustvarite
ustrezen razred in mu dodajte metode:
zacetno_stanje … višina vode, ki je trenutno v posodi
dolij … količina vode, ki jo dolijemo v posodo (v litrih)
preliv … količina vode, ki izteče ob prekomernem dolivanju
stanje … trenutna višina vode v cisterni
Prilagodite izpise.
"""

import math

class Cisterna:
    def __init__(self, premer_m: float, visina_m: float):
        if premer_m <= 0 or visina_m <= 0:
            raise ValueError("Premer in višina morata biti pozitivna.")
        self.premer_m = premer_m
        self.visina_m = visina_m

        self._visina_vode_m = 0.0
        self._preliv_l = 0.0

    @property #da izpiše errorje tam ko je treba (da se ne ustavi pr errorju ampak gre naprej)
    def _povrsina_osnove_m2(self) -> float:
        r = self.premer_m / 2
        return math.pi * r * r

    def zacetno_stanje(self, visina_vode_m: float):
        if visina_vode_m < 0:
            raise ValueError("Začetna višina vode ne sme biti negativna.")
        if visina_vode_m > self.visina_m:
            odvec_m = visina_vode_m - self.visina_m
            odvec_l = odvec_m * self._povrsina_osnove_m2 * 1000
            self._preliv_l += odvec_l
            self._visina_vode_m = self.visina_m
            print(f"Začetno stanje presega cisterno. Prelilo je {odvec_l:.1f} L.")
        else:
            self._visina_vode_m = visina_vode_m
            print(f"Začetno stanje nastavljeno: {self._visina_vode_m:.3f} m vode.")

    def dolij(self, kolicina_l: float):
        if kolicina_l < 0:
            raise ValueError("Količina dolivanja ne sme biti negativna.")

        dodan_volumen_m3 = kolicina_l / 1000  
        dodana_visina_m = dodan_volumen_m3 / self._povrsina_osnove_m2

        nova_visina = self._visina_vode_m + dodana_visina_m

        if nova_visina <= self.visina_m:
            self._visina_vode_m = nova_visina
            print(f"Dolito: {kolicina_l:.1f} L. Nova višina vode: {self._visina_vode_m:.3f} m.")
        else:
            prekoracitev_m = nova_visina - self.visina_m
            preliv_l = prekoracitev_m * self._povrsina_osnove_m2 * 1000

            self._visina_vode_m = self.visina_m
            self._preliv_l += preliv_l

           
            dejansko_notri_l = kolicina_l - preliv_l
            if dejansko_notri_l < 0:  
                dejansko_notri_l = 0.0

            print(
                f"Dolito: {kolicina_l:.1f} L. Cisterna je polna.\n"
                f"V cisterni je ostalo: {dejansko_notri_l:.1f} L, prelilo: {preliv_l:.1f} L."
            )

    def preliv(self) -> float:
        """Skupna količina preliva (litri)."""
        return self._preliv_l

    def stanje(self) -> float:
        """Trenutna višina vode (m)."""
        return self._visina_vode_m

    def __str__(self):
        return (
            f"Cisterna(premer={self.premer_m} m, višina={self.visina_m} m) | "
            f"voda={self._visina_vode_m:.3f} m | preliv={self._preliv_l:.1f} L"
        )


# Primer uporabe:
if __name__ == "__main__":
    c = Cisterna(premer_m=1.2, visina_m=2.0)
    c.zacetno_stanje(0.5)
    c.dolij(300)  
    c.dolij(2000)  
    print("Preliv skupaj:", c.preliv(), "L")
    print("Stanje (višina):", c.stanje(), "m")
    print(c)
```

```python
"""
Imamo parkirno garažo, ki ima določeno kapaciteto avtomobilov in
trenutno število parkiranih avtomobilov. Uporabi razrede in metode za
naslednje funkcionalnosti:
● zacetno_stanje … začetno število avtomobilov v garaži
● vstop … število avtomobilov, ki želijo vstopiti v garažo
● odhod … število avtomobilov, ki zapustijo garažo
● stanje … trenutno število avtomobilov v garaži
Prilagodite izpise.
"""

class ParkirnaGaraza:
    def __init__(self, kapaciteta: int):
        if kapaciteta <= 0:
            raise ValueError("Kapaciteta mora biti pozitivno celo število.")
        self.kapaciteta = kapaciteta
        self._parkirani = 0

    def zacetno_stanje(self, st_avtov: int):
        if st_avtov < 0:
            raise ValueError("Začetno število avtov ne sme biti negativno.")

        if st_avtov > self.kapaciteta:
            self._parkirani = self.kapaciteta
            visek = st_avtov - self.kapaciteta
            print(
                f"Začetno stanje presega kapaciteto. "
                f"V garaži je {self._parkirani} avtov, {visek} jih ni moglo parkirati."
            )
        else:
            self._parkirani = st_avtov
            print(f"Začetno stanje: v garaži je {self._parkirani} avtov.")

    def vstop(self, st_avtov: int):
        if st_avtov < 0:
            raise ValueError("Število avtov za vstop ne sme biti negativno.")

        prosta_mesta = self.kapaciteta - self._parkirani

        if st_avtov <= prosta_mesta:
            self._parkirani += st_avtov
            print(f"Vstopilo je {st_avtov} avtov. Zdaj je v garaži {self._parkirani} avtov.")
        else:
            self._parkirani = self.kapaciteta
            zavrnjeni = st_avtov - prosta_mesta
            print(
                f"Vstopiti želi {st_avtov} avtov, prostora je za {prosta_mesta}.\n"
                f"Vstopilo: {st_avtov - zavrnjeni}, zavrnjeni: {zavrnjeni}. Garaža je polna ({self.kapaciteta})."
            )

    def odhod(self, st_avtov: int):
        if st_avtov < 0:
            raise ValueError("Število avtov za odhod ne sme biti negativno.")

        if st_avtov <= self._parkirani:
            self._parkirani -= st_avtov
            print(f"Odhod: {st_avtov} avtov. Zdaj je v garaži {self._parkirani} avtov.")
        else:
            dejanski_odhodi = self._parkirani
            self._parkirani = 0
            print(
                f"Odhod: želelo je oditi {st_avtov} avtov, a v garaži jih je bilo le {dejanski_odhodi}.\n"
                f"Vsi so odšli. Garaža je prazna."
            )

    def stanje(self) -> int:
        return self._parkirani

    def __str__(self):
        return f"ParkirnaGaraža(kapaciteta={self.kapaciteta}) | trenutno={self._parkirani}"


# Primer uporabe:
if __name__ == "__main__":
    g = ParkirnaGaraza(kapaciteta=50)
    g.zacetno_stanje(10)
    g.vstop(30)
    g.vstop(20)  
    g.odhod(15)
    g.odhod(100) 
    print("Trenutno stanje:", g.stanje())
    print(g)
```
