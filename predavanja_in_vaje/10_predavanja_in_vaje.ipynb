```python
"""
Doma imamo cisterno za vodo za zalivanje vrta. Cisterna je v obliki valja
in izmerimo ji lahko premer ter višino. S pomočjo razredov ustvarite
ustrezen razred in mu dodajte metode:
zacetno_stanje … višina vode, ki je trenutno v posodi
dolij … količina vode, ki jo dolijemo v posodo (v litrih)
preliv … količina vode, ki izteče ob prekomernem dolivanju
stanje … trenutna višina vode v cisterni
Prilagodite izpise.
"""

import math

class Cisterna:
    def __init__(self, premer_m: float, visina_m: float):
        if premer_m <= 0 or visina_m <= 0:
            raise ValueError("Premer in višina morata biti pozitivna.")
        self.premer_m = premer_m
        self.visina_m = visina_m

        self._visina_vode_m = 0.0
        self._preliv_l = 0.0

    @property #da izpiše errorje tam ko je treba (da se ne ustavi pr errorju ampak gre naprej)
    def _povrsina_osnove_m2(self) -> float:
        r = self.premer_m / 2
        return math.pi * r * r

    def zacetno_stanje(self, visina_vode_m: float):
        if visina_vode_m < 0:
            raise ValueError("Začetna višina vode ne sme biti negativna.")
        if visina_vode_m > self.visina_m:
            odvec_m = visina_vode_m - self.visina_m
            odvec_l = odvec_m * self._povrsina_osnove_m2 * 1000
            self._preliv_l += odvec_l
            self._visina_vode_m = self.visina_m
            print(f"Začetno stanje presega cisterno. Prelilo je {odvec_l:.1f} L.")
        else:
            self._visina_vode_m = visina_vode_m
            print(f"Začetno stanje nastavljeno: {self._visina_vode_m:.3f} m vode.")

    def dolij(self, kolicina_l: float):
        if kolicina_l < 0:
            raise ValueError("Količina dolivanja ne sme biti negativna.")

        dodan_volumen_m3 = kolicina_l / 1000  
        dodana_visina_m = dodan_volumen_m3 / self._povrsina_osnove_m2

        nova_visina = self._visina_vode_m + dodana_visina_m

        if nova_visina <= self.visina_m:
            self._visina_vode_m = nova_visina
            print(f"Dolito: {kolicina_l:.1f} L. Nova višina vode: {self._visina_vode_m:.3f} m.")
        else:
            prekoracitev_m = nova_visina - self.visina_m
            preliv_l = prekoracitev_m * self._povrsina_osnove_m2 * 1000

            self._visina_vode_m = self.visina_m
            self._preliv_l += preliv_l

           
            dejansko_notri_l = kolicina_l - preliv_l
            if dejansko_notri_l < 0:  
                dejansko_notri_l = 0.0

            print(
                f"Dolito: {kolicina_l:.1f} L. Cisterna je polna.\n"
                f"V cisterni je ostalo: {dejansko_notri_l:.1f} L, prelilo: {preliv_l:.1f} L."
            )

    def preliv(self) -> float:
        """Skupna količina preliva (litri)."""
        return self._preliv_l

    def stanje(self) -> float:
        """Trenutna višina vode (m)."""
        return self._visina_vode_m

    def __str__(self):
        return (
            f"Cisterna(premer={self.premer_m} m, višina={self.visina_m} m) | "
            f"voda={self._visina_vode_m:.3f} m | preliv={self._preliv_l:.1f} L"
        )


# Primer uporabe:
if __name__ == "__main__":
    c = Cisterna(premer_m=1.2, visina_m=2.0)
    c.zacetno_stanje(0.5)
    c.dolij(300)  
    c.dolij(2000)  
    print("Preliv skupaj:", c.preliv(), "L")
    print("Stanje (višina):", c.stanje(), "m")
    print(c)
```

```python
"""
Imamo parkirno garažo, ki ima določeno kapaciteto avtomobilov in
trenutno število parkiranih avtomobilov. Uporabi razrede in metode za
naslednje funkcionalnosti:
● zacetno_stanje … začetno število avtomobilov v garaži
● vstop … število avtomobilov, ki želijo vstopiti v garažo
● odhod … število avtomobilov, ki zapustijo garažo
● stanje … trenutno število avtomobilov v garaži
Prilagodite izpise.
"""

class ParkirnaGaraza:
    def __init__(self, kapaciteta: int):
        if kapaciteta <= 0:
            raise ValueError("Kapaciteta mora biti pozitivno celo število.")
        self.kapaciteta = kapaciteta
        self._parkirani = 0

    def zacetno_stanje(self, st_avtov: int):
        if st_avtov < 0:
            raise ValueError("Začetno število avtov ne sme biti negativno.")

        if st_avtov > self.kapaciteta:
            self._parkirani = self.kapaciteta
            visek = st_avtov - self.kapaciteta
            print(
                f"Začetno stanje presega kapaciteto. "
                f"V garaži je {self._parkirani} avtov, {visek} jih ni moglo parkirati."
            )
        else:
            self._parkirani = st_avtov
            print(f"Začetno stanje: v garaži je {self._parkirani} avtov.")

    def vstop(self, st_avtov: int):
        if st_avtov < 0:
            raise ValueError("Število avtov za vstop ne sme biti negativno.")

        prosta_mesta = self.kapaciteta - self._parkirani

        if st_avtov <= prosta_mesta:
            self._parkirani += st_avtov
            print(f"Vstopilo je {st_avtov} avtov. Zdaj je v garaži {self._parkirani} avtov.")
        else:
            self._parkirani = self.kapaciteta
            zavrnjeni = st_avtov - prosta_mesta
            print(
                f"Vstopiti želi {st_avtov} avtov, prostora je za {prosta_mesta}.\n"
                f"Vstopilo: {st_avtov - zavrnjeni}, zavrnjeni: {zavrnjeni}. Garaža je polna ({self.kapaciteta})."
            )

    def odhod(self, st_avtov: int):
        if st_avtov < 0:
            raise ValueError("Število avtov za odhod ne sme biti negativno.")

        if st_avtov <= self._parkirani:
            self._parkirani -= st_avtov
            print(f"Odhod: {st_avtov} avtov. Zdaj je v garaži {self._parkirani} avtov.")
        else:
            dejanski_odhodi = self._parkirani
            self._parkirani = 0
            print(
                f"Odhod: želelo je oditi {st_avtov} avtov, a v garaži jih je bilo le {dejanski_odhodi}.\n"
                f"Vsi so odšli. Garaža je prazna."
            )

    def stanje(self) -> int:
        return self._parkirani

    def __str__(self):
        return f"ParkirnaGaraža(kapaciteta={self.kapaciteta}) | trenutno={self._parkirani}"


# izpisi
if __name__ == "__main__":
    g = ParkirnaGaraza(kapaciteta=50)
    g.zacetno_stanje(10)
    g.vstop(30)
    g.vstop(20)  
    g.odhod(15)
    g.odhod(100) 
    print("Trenutno stanje:", g.stanje())
    print(g)
```

```python
"""
V spletni učilnici (vaje) je datoteka zaposleni.csv.
Naloge:
a) preglejte podatke in ustvarite ustrezno podatkovno strukturo seznam
razredov
b) odprite datoteko in napolnite podatkovno strukturo s podatki
c) izpišite število programerjev v podjetju in število zaposlenih v IT odd.
d) izpišite koliko je najvišja plača v podjetju in kdo jo prejema?
e) izpišite koliko znaša povprečna plača v posameznem oddelku?
f) izpišite najvišjo plačo po delovnih mestih
g) izpišite koliko zaposlenih je po posamezni stopnji izobrazbe.
Preverite odgovore z Excelom.
"""

class Zaposleni:
    def __init__(self, id, priimek, ime, naslov, posta, kraj,
                 datum_rojstva, spol, izobrazba, oddelek,
                 delovno_mesto, datum_zaposlitve, placa):
        self.id = int(id)
        self.priimek = priimek
        self.ime = ime
        self.naslov = naslov
        self.posta = posta
        self.kraj = kraj
        self.datum_rojstva = datum_rojstva
        self.spol = spol
        self.izobrazba = izobrazba
        self.oddelek = oddelek
        self.delovno_mesto = delovno_mesto
        self.datum_zaposlitve = datum_zaposlitve
        self.placa = float(placa) 

    def __str__(self):
        return f"{self.ime} {self.priimek} | {self.oddelek} | {self.delovno_mesto} | {self.placa:.2f} €"


def parse_placa(v): #da damo plačo v cifro iz stringa
    s = str(v).replace("€", "").strip()
    s = s.replace(".", "")     
    s = s.replace(",", ".")    
    if s == "":
        return 0.0
    return float(s)


# b)
def preberi_zaposlene(pot):
    zaposleni = []

    f = open(pot, "r", encoding="utf-8")
    lines = f.readlines()
    f.close()

    header = lines[0].strip().split(";")

    for line in lines[1:]:
        line = line.strip()
        if line == "":
            continue

        parts = line.split(";")
        if len(parts) != 13:
            continue

        (id_, priimek, ime, naslov, posta, kraj, datum_rojstva, spol,
         izobrazba, oddelek, delovno_mesto, datum_zaposlitve, placa_raw) = parts

        placa = parse_placa(placa_raw)

        z = Zaposleni(id_, priimek, ime, naslov, posta, kraj,
                      datum_rojstva, spol, izobrazba, oddelek,
                      delovno_mesto, datum_zaposlitve, placa)
        zaposleni.append(z)

    return zaposleni


# c)
def stevilo_programerjev(zs):
    st = 0
    for z in zs:
        if "programer" in z.delovno_mesto.lower():
            st += 1
    return st

def stevilo_it_oddelek(zs):
    st = 0
    for z in zs:
        if "it" in z.oddelek.lower():
            st += 1
    return st


# d)
def najvisja_placa(zs):
    naj = zs[0]
    for z in zs[1:]:
        if z.placa > naj.placa:
            naj = z
    return naj


# e)
def povprecje_po_oddelkih(zs):
    vsote = {}   
    stevci = {} 

    for z in zs:
        od = z.oddelek
        if od not in vsote:
            vsote[od] = 0.0
            stevci[od] = 0
        vsote[od] += z.placa
        stevci[od] += 1

    povp = {}
    for od in vsote:
        povp[od] = vsote[od] / stevci[od]
    return povp


# f)
def najvisja_po_delovnih_mestih(zs):
    max_po_dm = {}
    for z in zs:
        dm = z.delovno_mesto
        if dm not in max_po_dm:
            max_po_dm[dm] = z.placa
        else:
            if z.placa > max_po_dm[dm]:
                max_po_dm[dm] = z.placa
    return max_po_dm


# g) 
def stevilo_po_izobrazbi(zs):
    st_po_iz = {}  
    for z in zs:
        iz = z.izobrazba
        if iz not in st_po_iz:
            st_po_iz[iz] = 0
        st_po_iz[iz] += 1
    return st_po_iz

#izpisi
if __name__ == "__main__":
    pot = r"C:/Users/tjasa/Desktop/FIS_Programiranje_1/predavanja_in_vaje/Uvod_v_Python/Zaposleni.csv"
    zs = preberi_zaposlene(pot)

    print("a) Podatkovna struktura: seznam razredov ->", type(zs), "z elementi tipa", type(zs[0]))
    print("b) Prebranih zaposlenih:", len(zs))

    print("\nc) Programerji:", stevilo_programerjev(zs))
    print("   Zaposleni v IT oddelku:", stevilo_it_oddelek(zs))

    naj = najvisja_placa(zs)
    print(f"\nd) Najvišja plača: {naj.placa:.2f} €")
    print("   Prejema jo:", naj.ime, naj.priimek, "|", naj.oddelek, "|", naj.delovno_mesto)

    print("\ne) Povprečna plača po oddelkih:")
    povp = povprecje_po_oddelkih(zs)
    for od in sorted(povp):
        print(f"   - {od}: {povp[od]:.2f} €")

    print("\nf) Najvišja plača po delovnih mestih:")
    mx_dm = najvisja_po_delovnih_mestih(zs)
    for dm in sorted(mx_dm):
        print(f"   - {dm}: {mx_dm[dm]:.2f} €")

    print("\ng) Število zaposlenih po izobrazbi:")
    po_iz = stevilo_po_izobrazbi(zs)
    for iz in sorted(po_iz):
        print(f"   - {iz}: {po_iz[iz]}")
```

```python
"""
Ustvarite razred bancni_racun, ki ima javno lastnost ID, privatne lastnosti ime, priimek, 
limit in stanje. Izdelajte objekt br1. Upoštevajte, da je začetni limit 300€. Naloge: 
a) nastavi vrednosti za ime, priimek, limit in stanje Napišite metodo, ki: 
b) izvede polog denarja na račun 
c) dvig denarja iz računa, ki hkrati preveri ali smo presegli limit. Če se to zgodi izvede dvig v vrednosti, ki je dovoljena in izpiše obvestilo 
d) izpiše stanje na računu z vsemi podatki o računu 
e) spremeni limit na novo vrednost f) prenos denarja 
iz računa br1 na br2. Potrebno je preveriti ali je znesek znotraj dovoljenih meja in nakazati največji možni znesek. Preverite odgovore z Excelom.
"""

class BancniRacun:
    def __init__(self, ID):
        self.ID = ID              
        self.__ime = ""           
        self.__priimek = ""       
        self.__limit = 300.0      
        self.__stanje = 0.0       

    # a)
    def nastavi_podatke(self, ime, priimek, limit=None, stanje=None):
        self.__ime = str(ime)
        self.__priimek = str(priimek)

        if limit is not None:
            self.spremeni_limit(limit)

        if stanje is not None:
            stanje = float(stanje)
            if stanje < -self.__limit:
                self.__stanje = -self.__limit
                print(f"[{self.ID}] Začetno stanje je prenizko. Nastavljeno na dovoljeno: {self.__stanje:.2f} €")
            else:
                self.__stanje = stanje

    def __max_dvig(self):
        return self.__stanje + self.__limit  

    # b) 
    def polog(self, znesek):
        znesek = float(znesek)
        if znesek <= 0:
            print(f"[{self.ID}] Polog mora biti > 0.")
            return 0.0

        self.__stanje += znesek
        print(f"[{self.ID}] Polog: {znesek:.2f} €. Novo stanje: {self.__stanje:.2f} €")
        return znesek

    # c)
    def dvig(self, znesek):
        znesek = float(znesek)
        if znesek <= 0:
            print(f"[{self.ID}] Dvig mora biti > 0.")
            return 0.0

        dovoljeno = self.__max_dvig()

        if znesek <= dovoljeno:
            self.__stanje -= znesek
            print(f"[{self.ID}] Dvig: {znesek:.2f} €. Novo stanje: {self.__stanje:.2f} €")
            return znesek
        else:
            if dovoljeno <= 0:
                print(f"[{self.ID}] Dvig ni mogoč. Dosežen limit (-{self.__limit:.2f} €).")
                return 0.0

            self.__stanje -= dovoljeno
            print(
                f"[{self.ID}] Zahtevan dvig {znesek:.2f} € presega limit.\n"
                f"        Izveden dvig največjega dovoljenega zneska: {dovoljeno:.2f} €.\n"
                f"        Novo stanje: {self.__stanje:.2f} €"
            )
            return dovoljeno

    # d) 
    def izpis(self):
        print("----- BANČNI RAČUN -----")
        print(f"ID:      {self.ID}")
        print(f"Ime:     {self.__ime}")
        print(f"Priimek: {self.__priimek}")
        print(f"Limit:   {self.__limit:.2f} €")
        print(f"Stanje:  {self.__stanje:.2f} €")
        print("------------------------")

    # e) 
    def spremeni_limit(self, nov_limit):
        nov_limit = float(nov_limit)
        if nov_limit < 0:
            print(f"[{self.ID}] Limit ne sme biti negativen.")
            return

        self.__limit = nov_limit

        if self.__stanje < -self.__limit:
            self.__stanje = -self.__limit
            print(f"[{self.ID}] Novi limit je nižji; stanje popravljeno na {self.__stanje:.2f} €")

        print(f"[{self.ID}] Limit nastavljen na {self.__limit:.2f} €")

    # f) 
    def prenos(self, znesek, prejemnik):
        znesek = float(znesek)
        if znesek <= 0:
            print(f"[{self.ID}] Prenos mora biti > 0.")
            return 0.0

        dovoljeno = self.__max_dvig()

        if dovoljeno <= 0:
            print(f"[{self.ID}] Prenos ni mogoč. Dosežen limit (-{self.__limit:.2f} €).")
            return 0.0

        dejanski = znesek if znesek <= dovoljeno else dovoljeno

        self.__stanje -= dejanski
        prejemnik.__stanje += dejanski  

        if dejanski < znesek:
            print(
                f"[{self.ID}] Zahtevan prenos {znesek:.2f} € presega dovoljeno mejo.\n"
                f"        Prenesen največji možni znesek: {dejanski:.2f} €"
            )
        else:
            print(f"[{self.ID}] Prenos uspešen: {dejanski:.2f} €")

        print(f"        Pošiljatelj stanje: {self.__stanje:.2f} € | Prejemnik stanje: {prejemnik.__stanje:.2f} €")
        return dejanski


# izpišemo:
if __name__ == "__main__":
    br1 = BancniRacun("BR1")
    br2 = BancniRacun("BR2")

    # a) 
    br1.nastavi_podatke("Tjaša", "Žilavec", limit=300, stanje=100)
    br2.nastavi_podatke("Aleš", "Frelih", limit=300, stanje=20)

    # b)
    br1.polog(50)

    # c) 
    br1.dvig(600)

    # d) 
    br1.izpis()
    br2.izpis()

    # e) 
    br1.spremeni_limit(200)

    # f) 
    br1.prenos(500, br2)

    br1.izpis()
    br2.izpis()
```
