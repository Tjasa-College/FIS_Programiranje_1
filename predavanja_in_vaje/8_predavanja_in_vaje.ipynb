```python
#c. koliko je prijavljenih tekmovalcev po posameznih kategorijah; nasvet: bodite pozorni, 
#da nekateri tekmovalci še niso prijavili kategorije

file = r"C:/Users/tjasa/Desktop/FIS_Programiranje_1/predavanja_in_vaje/Uvod_v_Python/Turnir.csv"

with open(file, "rt", encoding="utf-8") as f:
    vrstice = [v.strip() for v in f.read().splitlines() if v.strip()]

glava = vrstice[0].split(";")
i_kat = glava.index("Kategorija")

podatki = [v.split(";") for v in vrstice[1:]]

kategorije = {}
for vrstica in podatki:
    if len(vrstica) <= i_kat:
        vrstica += [""] * (i_kat - len(vrstica) + 1)

    kat = vrstica[i_kat].strip()
    if kat == "":
        kat = "ni_prijavljen"

    kategorije[kat] = kategorije.get(kat, 0) + 1

print("Število tekmovalcev po kategorijah:")
for kat in sorted(kategorije, key=lambda k: (-kategorije[k], k)):
    print(f"{kat}: {kategorije[kat]}")

```

```python
#Napišite program v Pythonu, ki izpiše:
#1. zahteva vnos naravnega števila in s pomočjo izjem preveri ali je uporabnik
#naredil napako pri vnosu in mu to napako v slovenskem jeziku tudi napiše.
#Možne napake so:
#- neštevilski vnos
#- negativno število ali število 0
#- decimalno število
#V primeru pravilnega vnosa izpišite vneseno vrednost

while True:
    vnos = input("Vnesi naravno število: ").strip()
    try:
        if vnos == "":
            raise ValueError("neštevilski vnos")
        if "." in vnos or "," in vnos:
            try:
                float(vnos.replace(",", "."))
                raise ValueError("decimalno število")
            except ValueError:
                raise ValueError("neštevilski vnos")
        n = int(vnos)
        if n <= 0:
            raise ValueError("negativno ali nič")
        print(f"Bravo: {n}")
        break

    except ValueError as e:
        msg = str(e)
        if msg == "decimalno število":
            print("Napaka: vnesli ste decimalno število.")
        elif msg == "negativno ali nič":
            print("Napaka: vnesli ste 0 ali negativno število.")
        else:
            print("Napaka: vnos ni število.")
```

```python
#3. Napišite program v Pythonu, ki izpiše koliko let, mesecev in dni je v Sloveniji evro plačilno sredstvo (1.1.2007). (Nasvet: uporabite datetime)

from datetime import date, timedelta

zacetek = date(2007, 1, 1)
danes = date.today()

leta = danes.year - zacetek.year
meseci = danes.month - zacetek.month
dni = danes.day - zacetek.day

if dni < 0:
    if danes.month == 1:
        prejsnji_mesec = date(danes.year - 1, 12, 1)
    else:
        prejsnji_mesec = date(danes.year, danes.month - 1, 1)
    zadnji_prejsnji = date(danes.year, danes.month, 1) - timedelta(days=1)

    dni += zadnji_prejsnji.day
    meseci -= 1

if meseci < 0:
    meseci += 12
    leta -= 1

print(f"Evro je v Sloveniji plačilno sredstvo že {leta} let, {meseci} mesecev in {dni} dni.")
print(f"(Od {zacetek} do {danes})")
```

```python
#4. Napišite program v Pythonu, ki izračuna koliko časa se je izvajal program, ki je 10 krat uredil po velikosti 100.000 naključnih števil v seznamu. (Nasvet: uporabite time)

import time
import random


zacetek = time.perf_counter()

for i in range(10):
    seznam = [random.randint(1, 1_000_000) for i in range(100_000)] #seznam random 100 000 števil med 1 in 1 000 000
    seznam.sort()

konec = time.perf_counter()
trajanje = konec - zacetek

print(f"Program je trajal {trajanje:.4f} sekund.")
```
