```python
def funkcija(a, b):
    try:
        c = (a + b ) / (a - b)
    except ZeroDivisionError:
        print("deljenje z nič")
    else:
        print(c)
    finally:
        print("konec izjeme")
funkcija (2, 3)
funkcija (3, 3)
```

```python
#Napišite program v Pythonu, ki izpiše:
#1. zahteva vnos naravnega števila in s pomočjo izjem preveri ali je uporabnik
#naredil napako pri vnosu in mu to napako v slovenskem jeziku tudi napiše.
#Možne napake so:
#- neštevilski vnos
#- negativno število ali število 0
#- decimalno število
#V primeru pravilnega vnosa izpišite vneseno vrednost

def funkcija():
    try:
        stevilo = int(input('Vnesi celo število: '))
        assert stevilo > 0
        #break
    except AssertionError:
        print("Število mora biti večje od 0")
    except FloatingPointError:
        print("Število je decimalno")
    except ValueError:
        print("Prišlo je do neštevilskega vnosa")
    else:
        print(stevilo)
    finally:
        print("Konec")


funkcija()
```
