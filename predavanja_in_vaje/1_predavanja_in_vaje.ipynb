**Primer: Program za pretvorbo temperature**

Program naj izvede naslednji postopek:</br></br>

1. Uporabnika naj vpraša po temperaturi v
stopinjah Celzija
2. Izpiše naj temperaturo v Kelvinih in Fahrenheitih

```python
#iz C v K: T(K) = T(°C) + 273.15
#iz C v F: T(°F) = T(C) × 9/5+32

temp_c = float(input("Vnesite temperaturo v stopinjah Celzija: "))

def temp_v_Kelvinih(temp):
    return temp + 273.15

def temp_v_Fahrenheit(temp):
    return (temp * 9 / 5) + 32


print(f"Temperatura {temp_c} v Celzijih je v Kelvinih {temp_v_Kelvinih(temp_c)} in v Fahrenheitih {temp_v_Fahrenheit(temp_c)}")

```
