```python
#Naloge:
#1. naloga: Kdo je zmagal (najmanjši skupni čas)?
tekmovalci = {
    "Ana":   {"stevilka": 1, "plavanje": 30, "kolesarjenje": 90, "tek": 45},
    "Bine":  {"stevilka": 2, "plavanje": 25, "kolesarjenje": 95, "tek": 50},
    "Cilka": {"stevilka": 3, "plavanje": 28, "kolesarjenje": 85, "tek": 48},
    "Domen": {"stevilka": 4, "plavanje": 33, "kolesarjenje": 88, "tek": 46},
    "Eva":   {"stevilka": 5, "plavanje": 29, "kolesarjenje": 92, "tek": 47},
}

zmagovalec = None
najmanjsi_cas = None

for ime, podatki in tekmovalci.items():
    skupaj = podatki["plavanje"] + podatki["kolesarjenje"] + podatki["tek"]
    if najmanjsi_cas is None or skupaj < najmanjsi_cas:
        najmanjsi_cas = skupaj
        zmagovalec = ime

print("Zmagovalec je:", zmagovalec)
print("Skupni čas (min):", najmanjsi_cas)
```

```python
#6. naloga: Izpišite rezultate za prve tri tekmovalce (dobitnike medalj) po vsaki disciplini.
tekmovalci = {
    "Ana":   {"stevilka": 1, "plavanje": 30, "kolesarjenje": 90, "tek": 45},
    "Bine":  {"stevilka": 2, "plavanje": 25, "kolesarjenje": 95, "tek": 50},
    "Cilka": {"stevilka": 3, "plavanje": 28, "kolesarjenje": 85, "tek": 48},
    "Domen": {"stevilka": 4, "plavanje": 33, "kolesarjenje": 88, "tek": 46},
    "Eva":   {"stevilka": 5, "plavanje": 29, "kolesarjenje": 92, "tek": 47},
}

discipline = ["plavanje", "kolesarjenje", "tek"]

for disc in discipline:
    top3 = sorted(tekmovalci.items(), key=lambda x: x[1][disc])[:3]

    print(f"\nTop 3 - {disc}:")
    for mesto, (ime, podatki) in enumerate(top3, start=1):
        print(f"{mesto}. {ime} (št. {podatki['stevilka']}): {podatki[disc]} min")
```
