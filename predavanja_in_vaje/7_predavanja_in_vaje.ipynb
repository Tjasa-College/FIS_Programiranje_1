```python
#/. razčleni čas v ure:minute

tekmovalci = {
    "Ana":   {"stevilka": 1, "plavanje": 30, "kolesarjenje": 90, "tek": 45},
    "Bine":  {"stevilka": 2, "plavanje": 25, "kolesarjenje": 95, "tek": 50},
    "Cilka": {"stevilka": 3, "plavanje": 28, "kolesarjenje": 85, "tek": 48},
    "Domen": {"stevilka": 4, "plavanje": 33, "kolesarjenje": 88, "tek": 46},
    "Eva":   {"stevilka": 5, "plavanje": 29, "kolesarjenje": 92, "tek": 47},
}

def v_ure_min(minute: int) -> str:
    ure = minute // 60
    min = minute % 60
    return f"{ure}:{min:02d}"   # npr. 1:05

for ime, p in tekmovalci.items():
    pl = v_ure_min(p["plavanje"])
    kol = v_ure_min(p["kolesarjenje"])
    tek = v_ure_min(p["tek"])
    print(f'{ime} (#{p["stevilka"]}): plavanje {pl}, kolesarjenje {kol}, tek {tek}')
```

```python
#V spletni učilnici je datoteka poste.txt. Prenesite jo v mapo
#Uvod_v_Python/datoteke. Zanima nas:

file = open ("C:/Users/tjasa/Desktop/FIS_Programiranje_1/predavanja_in_vaje/Uvod_v_Python/poste.txt", "rt", encoding="utf-8")
vrstice = file.readlines()
file.close()

skupine = {}         
imena_post = [] 

for vrstica in vrstice:
    vrstica = vrstica.strip()
    if vrstica == "":
        continue

    deli = vrstica.split()
    posta_txt = deli[0]

    if not (posta_txt.isdigit() and len(posta_txt) == 4):
        continue

    posta = int(posta_txt)
    skupina = (posta // 1000) * 1000
    skupine[skupina] = skupine.get(skupina, 0) + 1

    ime = " ".join(deli[1:])  
    imena_post.append(ime)

#a. koliko je posameznih poštnih številk v skupini 1000, 2000,...
for k in sorted(skupine):
    print(f"{k}: {skupine[k]}")

# b. katera skupina ima največ in katera najmanj poštnih številk
najvec_skupina = max(skupine, key=skupine.get)
najmanj_skupina = min(skupine, key=skupine.get)

print(f"Največ ima skupina {najvec_skupina} ({skupine[najvec_skupina]})")
print(f"Najmanj ima skupina {najmanj_skupina} ({skupine[najmanj_skupina]})")

# c. koliko imen pošt se prične na črko P
st_p = 0
for ime in imena_post:
    if ime.startswith("P"):
        st_p += 1
print("Število pošt, ki se začnejo na P:", st_p)

# d. koliko pošt ima ime iz več kot 1 besede
st_vec_besed = 0
for ime in imena_post:
    if len(ime.split()) > 1:
        st_vec_besed += 1
print("Število pošt z imenom iz več kot 1 besede:", st_vec_besed)
```
